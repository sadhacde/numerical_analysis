## Taylor Series Python Assignment

The $N$th order Taylor series formula is:
$$ f(x+h) \approx f(x) + \frac{h^1}{1!} f^{(1)}(x) +  \frac{h^2}{2!} f^{(2)}(x) + \ldots + \frac{h^N}{N!} f^{(N)}(x),$$
where $f^{(j)}(x)$ denotes the $j$th derivative of the function $f$ at the point $x$.

The error in the formula will be less than the Taylor error bound, which is given by:

$$ R_N(x+h) = \frac{|h|^{N+1}}{(N+1)!}\max_{y \in [x,x+h]}| f^{(N+1)}(y)|.$$ 

```python
#Warm-up exercise: Complete the following code that computes and plots 
# the 7th order Taylor series approximation and Taylor error bound for 
# f(x) = cos(x) with expansion point x=0 on the interval [-3, 3] 
# with increment 0.01.  
#
### Import section
# 3 necessary statements for numerical plotting in Jupyter
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# Factorial comes from 'math' package
from math import factorial
#
### Initialize section
#array of derivatives at 0 compute by hand
deriv = np.array([1, 0, -1, 0, 1, 0, -1, 0])
# degree of Taylor polynomial
N = len(deriv) - 1
# Max abs. value of N+1 deriv on interval (Compute using calculus)
# (used in error formula:  error <= abs(h^(N+1)/(N+1)) * max(abs(N+1th deriv.))
maxAbsNplus1deriv = 1
#array with h values
h = np.arange(-3,3.05,0.01)
# Give the expansion point for the Taylor series
expansion = 0
# Create an array to contain Taylor series
taylor = np.zeros(len(h))
#
### Calculation section
#calculate Taylor series for all h at once using a loop
for n in range(N+1):
    taylor = taylor + (h**n)/factorial(n)*deriv[n]

#actual function values 
actualFn = np.cos(expansion + h)
#error actual and taylor series
actualAbsError = abs(actualFn-taylor)
#theoretical error=|h^(deg+1)/(deg+1)!*max|n+1 deriv||
theoAbsError = abs(h**(N+1)/factorial(N+1)*maxAbsNplus1deriv)
#Compute actual x values (displaced by expansion point)
xActual = expansion + h
#
### Output section
#plot figures for cos and errors using
# Object-oriented style plotting
# Two subplots in a row
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
# objects for the left subplot (attached to axes #0)
axes[0].plot(xActual, taylor, 'r', label='Taylor series')
axes[0].plot(xActual, actualFn, 'b--', label='cos(x)')
axes[0].legend(loc = 2) #legend in upper left
axes[0].set_title('Taylor Series for cos(x)')
# objects for the right subplot (attached to axes #1)
axes[1].plot(xActual, actualAbsError, label='Actual Absolute Error')
axes[1].plot(xActual, theoAbsError, label='Theoretical Absolute Error')
axes[1].legend(loc = 2) #legend in upper left
axes[1].set_title('Actual and Theoretical Error')



```

```python
# exercise 1: 10th order Taylor series approximation and theoretical error of 𝑓(𝑥) = cos(3𝑥)
# expansion point 𝑥 = 0.5, on the interval [−0.5,2.5] with h-increment 0.1

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import factorial

deriv = np.array([0.070737, -2.992185, -0.636635, 26.932365, 5.729713, -242.391282, -51.567426, 2181.521536, 464.10678, -19633.603821, -4176.961021])
N = len(deriv) - 1
maxAbsNplus1deriv = 3**11 #11th

expansion = 0.5
taylor = np.zeros(len(h))
h = np.arange(-1.0, 2.01, 0.1)

for n in range(N+1):
    taylor = taylor + (h**n)/factorial(n)*deriv[n]
    
actualFn = np.cos(3 * (h + expansion))

actualAbsError = abs(actualFn - taylor)

#|h^(deg+1)/(deg+1)! * max|n+1 deriv||
theoAbsError = abs((h)**(N + 1) / factorial(N + 1) * maxAbsNplus1deriv)
xActual = expansion + h

# output section
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
axes[0].plot(xActual , taylor, 'r', label='Taylor series')
axes[0].plot(xActual , actualFn, 'b--', label='cos(3x)')
axes[0].legend(loc=3)
axes[0].set_title('Taylor Series for cos(3x)')

axes[1].plot(xActual , actualAbsError, label='Actual Absolute Error')
axes[1].plot(xActual , theoAbsError, label='Theoretical Absolute Error')
axes[1].legend(loc=2)
axes[1].set_title('Actual & Theoretical Error')
```

```python
# execise 2: 6th order Taylor series approximation and theoretical error for 𝑓(𝑥) = 𝑙𝑛(𝑥)
# expansion point 𝑥 = 1, on the interval [0.5,2] with h-increment 0.05

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import factorial, log

deriv = np.array([0, 1, -1, 2, -6, 24, -120])
N = len(deriv) - 1

maxAbsNplus1deriv = 720 #7th

h = np.arange(-0.5, 1.01, 0.05)

expansion = 1

taylor = np.zeros(len(h))

for n in range(N + 1):
    taylor += ((h**n) / factorial(n) * deriv[n])

actualFn = np.log((h + expansion))

actualAbsError = abs(taylor - actualFn)
theoAbsError = abs(h**(N + 1) / factorial(N + 1) * maxAbsNplus1deriv)
xActual = expansion + h

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

# plot
axes[0].plot(xActual, taylor, 'r', label='Taylor series')
axes[0].plot(xActual, actualFn, 'b--', label='ln(x)')
axes[0].legend(loc=2)
axes[0].set_title('Taylor Series for ln(x)')

# actual and theoretical absolute errors
axes[1].plot(xActual, actualAbsError, label='Actual Absolute Error')
axes[1].plot(xActual, theoAbsError, label='Theoretical Absolute Error')
axes[1].legend(loc=2)
axes[1].set_title('Actual & Theoretical Error')
```
